# Foundation Colors — Silver Layer

Reads current color records from `lego_brickbase.bronze.raw_colors` and loads them into the `lego_brickbase.silver.foundation_colors` Delta table.

## What this notebook does

1. **Read** — Selects all current, non-deleted records from the bronze `raw_colors` table.
2. **Transform** — Renames `id` to `color_key`, retains business columns (`id`, `name`, `rgb`, `is_trans`, `valid_from`), and derives a `valid_from_date` column used for partitioning.
3. **Write** — Overwrites the Delta table at `/Volumes/lego_brickbase/silver/delta_volume/colors`, partitioned by `valid_from_date`.
4. **Register** — Creates the catalog table `lego_brickbase.silver.foundation_colors` in Unity Catalog if it does not already exist.

## Imports and Configuration

Import Spark functions needed for the transformation. `BRONZE_TABLE` is the source; `DELTA_TABLE_PATH` is the physical Delta location on the Unity Catalog external volume; `CATALOG_TABLE` is the Unity Catalog table reference.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date


BRONZE_TABLE     = "lego_brickbase.bronze.raw_colors"
DELTA_TABLE_PATH = "/Volumes/lego_brickbase/silver/delta_volume/colors"
CATALOG_TABLE    = "lego_brickbase.silver.foundation_colors"
```

## Read and Transform

Filter the bronze table to current, non-deleted records, then rename `id` to `color_key` and derive `valid_from_date` (the date portion of `valid_from`) for use as the Delta partition column.

```python
df_source = (
    spark.table(BRONZE_TABLE)
    .filter((col("is_current") == True) & (col("is_deleted") == False))
)

df_foundation = (
    df_source
    .select(
        col("id").alias("color_key"),
        col("id").alias("color_id"),
        col("name").alias("color_name"),
        col("rgb"),
        col("is_trans").alias("is_transperent"),
        col("valid_from"),
    )
    .withColumn("valid_from_date", to_date(col("valid_from")))
)

df_foundation.printSchema()
df_foundation.display(10, truncate=False)
```

## Write to Delta Volume and Register Catalog Table

Overwrite the Delta table at the silver volume path, partitioned by `valid_from_date`. Then register it in Unity Catalog as `lego_brickbase.silver.foundation_colors` if it does not already exist.

```python
(
    df_foundation
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .partitionBy("valid_from_date")
    .save(DELTA_TABLE_PATH)
)

row_count = spark.read.format("delta").load(DELTA_TABLE_PATH).count()
print(f"Rows written to Delta table: {row_count:,}")

spark.sql(f"""
    CREATE OR REPLACE TABLE {CATALOG_TABLE}
    AS SELECT * FROM delta.`{DELTA_TABLE_PATH}`
""")
print(f"Catalog table ready: {CATALOG_TABLE}")
```
